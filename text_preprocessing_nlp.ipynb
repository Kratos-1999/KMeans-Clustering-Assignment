```python
import nltk
import string
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
```

```python
text = """
Artificial Intelligence (AI) is transforming the world rapidly! Machines can learn from data, recognize patterns, and make decisions.
However, many people ask: will AI replace humans? Experts say it will assist rather than replace most jobs.

Data science, machine learning, and natural language processing are powerful technologies.
They help organizations analyze information, predict trends, and improve decision-making.
But challenges remain: privacy, ethics, and transparency must be addressed!
"""
```

```python
tokens = word_tokenize(text)

print("Tokens:")
print(tokens)
```

```python
tokens_no_punct = [word for word in tokens if word not in string.punctuation]

print("After Removing Punctuation:")
print(tokens_no_punct)
```

```python
stop_words = set(stopwords.words('english'))

filtered_words = [word for word in tokens_no_punct if word.lower() not in stop_words]

print("After Stopword Removal:")
print(filtered_words)
```

```python
stemmer = PorterStemmer()

stemmed_words = [stemmer.stem(word) for word in filtered_words]

print("Stemmed Words:")
print(stemmed_words)
```

```python
lemmatizer = WordNetLemmatizer()

lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_words]

print("Lemmatized Words:")
print(lemmatized_words)
```

```python
print("Original Words:", filtered_words)
print("\nStemmed Words:", stemmed_words)
print("\nLemmatized Words:", lemmatized_words)
```
